```python
# ================================================
# SimulaciÃ³n: Flujo S3 -> EventBridge -> Lambda
# Autor: Juan Pablo Mora Diaz (ejercicio de arquitectura serverless)
# ================================================

import json
import logging
from types import SimpleNamespace

# Configurar logging
logger = logging.getLogger()
logger.setLevel(logging.INFO)

# --------------------------------
# FunciÃ³n Lambda simulada (igual a la de AWS)
# --------------------------------
def lambda_handler(event, context):
    logger.info("Evento recibido: %s", json.dumps(event))

    # Extraer detalles del evento simulado
    detail = event.get('detail', {})
    bucket = detail.get('bucket', {}).get('name')
    key = detail.get('object', {}).get('key')

    if bucket and key:
        print(f"âœ… Archivo cargado: {key}")
    else:
        print("âš ï¸ No se encontrÃ³ informaciÃ³n del archivo en el evento.")

    return {
        'statusCode': 200,
        'body': json.dumps('Evento procesado correctamente')
    }

# --------------------------------
# Evento simulado (como lo enviarÃ­a AWS EventBridge)
# --------------------------------
fake_event = {
    "version": "0",
    "id": "abc12345",
    "detail-type": "Object Created",
    "source": "aws.s3",
    "account": "123456789012",
    "time": "2025-10-22T14:30:00Z",
    "region": "us-east-1",
    "resources": ["arn:aws:s3:::eventbridge-demo-bucket"],
    "detail": {
        "bucket": {"name": "eventbridge-demo-bucket"},
        "object": {"key": "ejemplo.txt"}
    }
}

# --------------------------------
# Ejecutar simulaciÃ³n
# --------------------------------
print("ðŸš€ Iniciando simulaciÃ³n del flujo S3 -> EventBridge -> Lambda...\n")

response = lambda_handler(fake_event, SimpleNamespace())

print("\n-----------------------------------")
print("Respuesta Lambda simulada:")
print(json.dumps(response, indent=2))
print("-----------------------------------")
```
