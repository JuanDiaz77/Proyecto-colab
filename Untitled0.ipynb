```python
# 5. Generación del código Terraform para el bootstrap
bootstrap_main = """\
provider "aws" {
  region = "us-east-1"
}

resource "aws_s3_bucket" "tf_state_bucket" {
  bucket = var.state_bucket_name

  tags = {
    Name        = "Terraform State Bucket"
    Environment = "bootstrap"
  }
}

resource "aws_dynamodb_table" "tf_locks" {
  name         = var.lock_table_name
  billing_mode = "PAY_PER_REQUEST"
  hash_key     = "LockID"

  attribute {
    name = "LockID"
    type = "S"
  }

  tags = {
    Name        = "Terraform Locks Table"
    Environment = "bootstrap"
  }
}
"""
with open("terraform-multienv/bootstrap/main.tf", "w") as f:
    f.write(bootstrap_main)

#Define:

#Provider AWS

#Bucket S3 → almacena el remote state

#Tabla DynamoDB → almacena state locks

#Esto evita que múltiples personas rompan el estado de Terraform al trabajar simultáneamente.
```

```python
#5.2 Archivo variables.tf

bootstrap_vars = """\
variable "state_bucket_name" {
  description = "Nombre del bucket S3 para almacenar el estado de Terraform"
  type        = string
}

variable "lock_table_name" {
  description = "Nombre de la tabla DynamoDB para los bloqueos de estado"
  type        = string
}
"""
with open("terraform-multienv/bootstrap/variables.tf", "w") as f:
    f.write(bootstrap_vars)

#Define variables reutilizables para los nombres del bucket y la tabla.
```

```python
#5.3 Archivo outputs.tf

bootstrap_outputs = """\
output "state_bucket_name" {
  value = aws_s3_bucket.tf_state_bucket.bucket
}

output "lock_table_name" {
  value = aws_dynamodb_table.tf_locks.name
}
"""
with open("terraform-multienv/bootstrap/outputs.tf", "w") as f:
    f.write(bootstrap_outputs)
```
